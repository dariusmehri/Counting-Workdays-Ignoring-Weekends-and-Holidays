```python
import pandas as pd
import numpy as np
import datetime as dt

pd.set_option('chained_assignment', None)

df = pd.read_csv("S:...Utility Complaints.csv", low_memory=False)

df['Date Entered'] = pd.to_datetime(df['Date Entered'])
df['Disposition Date'] = pd.to_datetime(df['Disposition Date'])

```

```python
def workdays(start, end):
    """Calculates days between two dates, excluding weekends and holidays"""
    days = np.busday_count(start, end )
    
    #HOLIDAYS
    #2016
    new_year16 = dt.date(2016, 1, 1)
    mlk16 = dt.date(2016, 1, 18)
    pres16 = dt.date(2016, 2, 15)
    mem16 = dt.date(2016, 5, 30)
    ind16 = dt.date(2016, 7, 4)
    ld16 = dt.date(2016, 9, 5)
    col16 = dt.date(2016, 10, 10)
    elec16 = dt.date(2016, 11, 8)
    vet16 = dt.date(2016, 11, 11)
    thank16 = dt.date(2016, 11, 24)
    christ16 = dt.date(2016, 12, 26)

    #2017
    new_year17 = dt.date(2017, 1, 2)
    mlk17 = dt.date(2017, 1, 16)
    pres17 = dt.date(2017, 2, 20)
    mem17 = dt.date(2017, 5, 29)
    ind17 = dt.date(2017, 7, 4)
    ld17 = dt.date(2017, 9, 4)
    col17 = dt.date(2017, 10, 9)
    elec17 = dt.date(2017, 11, 7)
    vet17 = dt.date(2017, 11, 10)
    thank17 = dt.date(2017, 11, 23)
    christ17 = dt.date(2017, 12, 25)
    
    #Correct for 2016
    if (new_year16 > start and new_year16 < end):
        days = days - 1
    if (mlk16 > start and mlk16 < end):
        days = days - 1
    if (pres16 > start and pres16 < end):
        days = days - 1
    if (mem16 > start and mem16 < end):
        days = days - 1
    if (ind16 > start and ind16 < end):
        days = days - 1
    if (ld16 > start and ld16 < end):
        days = days - 1
    if (col16 > start and col16 < end):
        days = days - 1
    if (elec16 > start and elec16 < end):
        days = days - 1
    if (vet16 > start and vet16 < end):
        days = days - 1
    if (thank16 > start and thank16 < end):
        days = days - 1
    if (christ16 > start and christ16 < end):
        days = days - 1
    
    #Correct for 2017
    if (new_year17 > start and new_year17 < end):
        days = days - 1
    if (mlk17 > start and mlk17 < end):
        days = days - 1
    if (pres17 > start and pres17 < end):
        days = days - 1
    if (mem17 > start and mem17 < end):
        days = days - 1
    if (ind17 > start and ind17 < end):
        days = days - 1
    if (ld17 > start and ld17 < end):
        days = days - 1
    if (col17 > start and col17 < end):
        days = days - 1
    if (elec17 > start and elec17 < end):
        days = days - 1
    if (vet17 > start and vet17 < end):
        days = days - 1
    if (thank17 > start and thank17 < end):
        days = days - 1
    if (christ17 > start and christ17 < end):
        days = days - 1
        
    return days


df["Response Time"] = ""

for i in range(0, len(df)):
    start = dt.date(df["Date Entered"][i].year, df["Date Entered"][i].month, df["Date Entered"][i].day )
    end = dt.date(df["Disposition Date"][i].year,df["Disposition Date"][i].month, df["Disposition Date"][i].day )

    days = workdays(start, end)
    
    df["Response Time"][i] = days

 
```

```python
df.to_csv("S:..Utility Complaints Response Time.csv")
```
